# Zmienne i operatory

## Zadanie 1

Zmienna `predkosc_ms` ma wartość prędkości w m/s.
Do zmiennej `predkosc_kmh` przypisz wartość prędkości w km/h.

Wartość zmiennej `predkosc_kmh` powinna zależeć od wartości zmiennej `predkosc_ms`.
Wzór na przeliczenie m/s na km/h znajdziesz poniżej.

$$v_{kmh} = v_{ms}\frac{3600}{1000}$$

````python
predkosc_ms = 120.5
predkosc_kmh = predkosc_ms * 3600 / 1000

print(str(predkosc_ms) + '[m/s] = ' + str(predkosc_kmh) + ' [km/h]')
````

## Zadanie 2

Należy sprawdzić, czy dla podanych wartości zmiennych `p` i `q` zachodzi równość
Wynik powinien znaleźć się w zmiennej `result`

$$ \neg p \vee \neg q = \neg(p \wedge q)$$

````python
p = True
q = False

result = not p or not q == (not (p and q))

print(result)
````
