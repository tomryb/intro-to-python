```python
def triangle_field(a, h):
    """
    Args:
        a: długość podstawy
        h: wysokość
    """
    return a * h * 0.5

print(triangle_field(9, 2))
```

Uzupełnij funkcję `circle_field` tak, aby zwracała pole koła dla podanej długości promienia. Wartość liczby pi dostępna jest w zmiennej `pi` w module `math`.

Wywołaj funkcję dla różnych parametrów i sprawdź jej działanie.

```python
import math
```

```python
def circle_field(r):
    """
    Args:
        r: promień okręgu
    """
    return math.pi * r**2
```

```python
print(circle_field(4))
print(circle_field(80))
```

Oczywiście najprostszym sposobem na rozwiązanie zadania byłoby użycie operatora potęgowania `**`, warunki zadania jednak tego zabraniają.

Chodzi o użycie pętli. Zaczynamy od wartości `result=1`, aby użyć elementu neutralnego dla mnożenia. Jednocześnie `range(0)` daje nam pustą sekwencję, co powoduje brak wykonania pętli i załatawia poprawny wynik dla potęg o wykładniku `0`. 



```python
def power(a, n):
    """
    Args:
        a: podstawa
        n: wykładnik
    """
    result = 1
    for i in range(n):
        result *= a
    return result

power(2, 0)
power(2, 1)
power(2, 2)
power(10, 2)
```
