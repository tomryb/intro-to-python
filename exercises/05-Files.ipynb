# Zadanie 1

Uzupełnij funkcję `word_stats` w taki sposób, aby zwracała ilość wystąpień słów w pliku o nazwie przekazanej w parametrze filename.

Wartością zwracaną powinien być słownik o nas†epującej strukturze:

`{'słowo1': ilość wystąpień, 'słowo2': ilość_wystąpień}`

Funkcja powinna porównywać słowa niezależnie od wielkości liter.

Funkcja powinna także oczyścić tekst ze znaków przestankowych oraz białych znaków.

## Uwagi

`linia 4`: Wczytuję zawartość pliku tekstowego do zmiennej (zawierającej zały tekst formie `str`)

`linia 5`: `text.split()` podzieli tekst na wyrazy i wytnie wszystkie białe znaki spomiędzy.

`linia 6`: Normalizuję słowa do zapisu małą literą. Metoda `.strip()` wytnie wszystkie znaki przestankowe (należące do `string.punctuation`) możemy wtedy dodać słowo do słownika statystyk wyrazów.

`linia 8`: `.setdefault()` Gwarantuje nam dodanie nieistniejącego klucza i inicjalizację wartością `0`.

```python
import string
```

```python
def word_stats(filename):
    stats = {}
    with open(filename, 'r') as f:
        text = f.read()
    for word in text.split():
        word = word.lower().strip(string.punctuation)
        if word != '':
            stats.setdefault(word, 0)
            stats[word] += 1
    return stats

pprint(word_stats(TEXT_FILE))
```

# Zadanie 3

Uzupełnij funkcję `unscramble` tak, aby zapisywała do nowego pliku zawartość pliku, którego nazwa zostaje przekazana jako argument wejściowy.
Przed zapisem zawartość powinna zostać zmodyfikowana wg. następującego algorytmu:

1. Odczytać 100 bajtów pliku.
2. Jeśli udało się odczytać 100 bajtów to do pliku wyjściowego zapisujemy najpierw ostatnie 50 bajtów z odczytanego bufora, następnie pierwsze 50 bajtów.
3. Jeśli odczytaliśmy mniej niż 100 bajtów (nie ma więcej danych w pliku) to zapisujemy do pliku wyjściowego odczytane dane bez zmian.
4. punkty 1-3 należy powtórzyć dla całego pliku.
 
Funkcja nie powinna zwracać żadnej wartości, ale powinna tworzyć plik wyjściowy, 
dla każdego pliku wejściowego. Nazwa pliku wyjściowego powinna być taka sama jak wejściowego z usuniętym prefiksem "scrambled_"

Przy poprawnym wykonaniu algorytmu pliki wyjściowe powinny być obrazami w formacie JPG.

Na którym z nich znajduje się zwierzę?

## Uwagi

Zdecydowanie najtrudniejsze zadanie z plików

`linia 2`: usunięcie prefiksu `scrambled` z nazwy pliku. Można to zrobić innym sposobem. Nie było sednem zadania, więc wykonałem prostą (i niezbyt odporną na błędy) operację.

`linie 3 i 4`: Otwarcie plików wejściowego i wyjściowego w trybie binarnym, odpowiednio do odczytu i zapisu. Tylko otwarcie binarne daje nam gwarancję odczytu bajt, po bajcie.

`linia 6`: Warunek kontunuacji odczytu, jeśli plik będzie pusty dostaniemy pusty obiekt `bytes`.

`linie 8 i 9`: Jako, że typ `bytes` jest ciągiem znaków można użyć operatora `slice`, aby uzyskać jego fragmenty i zapisać w odwrotnej koleności przy pomocy metody `.write()`.

Funkcja nie zwraca wartości, a jedynie generuje pliki wyjściowe, które potem można podejrzeć dowolną przeglądarką obrazków.

```python
def unscramble(filename):
    out_filename = filename.lstrip('scrambled_')
    with open(filename, 'rb') as in_file:
        with open(out_filename, 'wb') as out_file:
            chunk = in_file.read(100)
            while len(chunk) > 0:
                if len(chunk) == 100:
                    out_file.write(chunk[50:])
                    out_file.write(chunk[:50])
                else:
                    out_file.write(chunk)
                chunk = in_file.read(100)
            
for filename in PICTURES:
    unscramble(filename)
```
