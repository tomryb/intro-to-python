# Zadanie 3

Funkcja `add_underscores` powinna działać podobnie do funkcji `append_underscores`.
Z tą różnicą, że znaki '_' powinny być równomiernie rozłożone przed i za słowem. Jeśli wymagane jest dodanie nieparzystej liczby znaków '_' to większa ich ilość powinna znajdować się za słowem.

Przykład:

## Uwagi

Zachodzą te same właściwości, które zostały wymienione w uwagach do zadania 2.
Dodatkowo, wykorzystuję dzielenie całkowite przez 2, aby w łatwy sposób uzyskać długość prefiksu oraz sufiksu, także dla nieparzystych ilości znaków.

Dodanie nadmiarowego znaku `_` z prawej strony przy nieparzystym wyniku osiągam przez sprawdzenie reszty z dzielenia wymaganej ilości znaków przez 2, która wynosić musi 0 lub 1, w związku z tym dla parzystej ilości otrzymam ciąg `'_'`, a dla nieparzystej `''`.

```python
words = ['Początek', 'traktatu', 'czasu', 'być', 'wstrzemięźliwym', 'Tak',
         'może', 'być', 'uważana', 'jako', 'najwyższy', 'stopień', 'moralności',
         'z', 'części', 'Ale', 'się', 'od', 'Stworzyciela', 'niebył',
         'nadany', 'Jakże', 'to', 'mogło', 'być', 'moralna', 'a', 'jednak',
         'złą', 'sprawę', 'będzie', 'znowu', 'się', 'uczynkiem', 'ale', 'każdy',
         'będzie', 'znowu', 'się', 'łaskawego', 'przyjęcia', 'żegnam', 'się',
         'on', 'nieraz', 'zgryzot', 'sumienia', 'które', 'przybieramy', 'z',
         'czystego', 'serca', 'lub', 'obowiąski', 'które', 'człowiek', 'jako',
         'najwyższy', 'stopień', 'przyjaźni', 'świadczyć', 'lecz', 'tylko',
         'robakiem', 'tedy', 'jej', 'człowiek', 'w', 'Królewcu', 'szanowny',
         'Imanuel', 'Kant', 'Professor', 'filozofii', 'który', 'się', 'musiał',
         'od', 'Stworzyciela', 'niebył', 'nadany']

def add_underscores(words):
    max_length = len(longest_word(words))
    out = []
    for word in words:
        prefix = '_' * ((max_length - len(word)) // 2)
        suffix = prefix + '_' * ((max_length - len(word)) % 2)
        out.append(f'{prefix}{word}{suffix}')
    return out

pprint(add_underscores(words))
```

## Uwaga

Wykorzystuję fakt, że jeśli słowo zaczyna się wielką literą to po użyciu na tym znaku metody `.upper()` wynik będzie taki sam jak bez jej użycia. Wystarczy więc proste porównanie, aby ocenić, czy mamy do czynienia ze słowem rozpoczynającym się wielką, czy małą literą.

```python
words = ['Początek', 'traktatu', 'czasu', 'być', 'wstrzemięźliwym', 'Tak',
         'może', 'być', 'uważana', 'jako', 'najwyższy', 'stopień', 'moralności',
         'z', 'części', 'Ale', 'się', 'od', 'Stworzyciela', 'niebył',
         'nadany', 'Jakże', 'to', 'mogło', 'być', 'moralna', 'a', 'jednak',
         'złą', 'sprawę', 'będzie', 'znowu', 'się', 'uczynkiem', 'ale', 'każdy',
         'będzie', 'znowu', 'się', 'łaskawego', 'przyjęcia', 'żegnam', 'się',
         'on', 'nieraz', 'zgryzot', 'sumienia', 'które', 'przybieramy', 'z',
         'czystego', 'serca', 'lub', 'obowiąski', 'które', 'człowiek', 'jako',
         'najwyższy', 'stopień', 'przyjaźni', 'świadczyć', 'lecz', 'tylko',
         'robakiem', 'tedy', 'jej', 'człowiek', 'w', 'Królewcu', 'szanowny',
         'Imanuel', 'Kant', 'Professor', 'filozofii', 'który', 'się', 'musiał',
         'od', 'Stworzyciela', 'niebył', 'nadany']

def count_capitals(words):
    count = 0
    for word in words:
        if word[0] == word[0].upper():
            count += 1
    return count
pprint(count_capitals(words))
```
