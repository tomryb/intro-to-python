# Instruckja warunkowa

```python
x = -10
```

```python
if x != 0:
    print('x nie jest zerem')
```

```python
if x < 0:
    print('x jest liczbą ujemną')
else:
    print('x jest liczbą nieujemną')
```

```python
if x < 0:
    print('x jest liczbą ujemną')
elif x == 0:
    print('x jest zerem')
else:
    print('x jest liczbą dodatnią')
```

# Pętla `for`

```python
items = ['a', 'b', 1, 'c', 'd', 2, 'e', 'f', False]
for i in items:
    print(i)
```

```python
for x in range(10):
    print(x)
```

```python
items = ['a', 'b', 1, 'c', 'd', 2, 'e', 'f', False]
for idx, value in enumerate(items):
    print(idx, value)
```

```python
items = ['a', 'b', 1, 'c', 'd', 2, 'e', 'f', False]
for i in items:
    if i == 'b':
        continue
    print(i)
```

```python
items = ['a', 'b', 1, 'c', 'd', 2, 'e', 'f', False]
for i in items:
    if i == 'b':
        break
    print(i)
```

# Pętla `while`

```python
x = 10
while x > 0:
    print(x)
    x -= 1 # x = x - 1
```
