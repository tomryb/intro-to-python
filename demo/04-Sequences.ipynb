## Odwracanie list

```python
lst = [1, 2, 3, 4, 5]
print(reversed(lst))
print(lst)

for x in reversed(lst):
    print(x)
```

```python
lst = [1, 2, 3, 4, 5]
print(lst)
lst.reverse()
print(lst)
```

```python
lst = [1, 2, 3, 4, 5]
reversed_lst = lst[::-1]
print(lst)
print(reversed_lst)
```

# Rozpakowywanie wartości

```python
t = (1, 2, 3)
x, y, z = t

print(x, y, z)
```

```python
x, y, z = (1, 2, 3)
x, y, z = 1, 2, 3
```

```python
x, y = 10, 11
print (x, y)
x, y = y, x
print (x, y)
```

```python
x, _, _, y = 1, 2, 3, 4
print(x, y)
```

# Indeksy oraz Slice

```python
lst = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

print(lst[2])
print(lst[0:5])
print(lst[1:])
print(lst[:-1])
print(lst[::-1])
print(lst[:])
```

# Operatory

```python
lst = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

print(len(lst))

print('a' in lst)

if 'b' in lst:
    print('b jest elementem listy')

if 'x' not in lst:
    print('x nie jest elementem listy')
```

# Iterowanie

```python
s = 'Ala ma kota'
for letter in s:
    print(letter)
```
