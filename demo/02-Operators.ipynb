# Operatory arytmetyczne

```python
x = 7 + 5
y = 7 + 4.5
print(x, y)
```

```python
x = 7 * 7
y = 7 * 7.5

print(x, y)
```

```python
x = 12 / 4
y = 9 / 4.5

print(x, y)
```

```python
x = 12 // 4
y = 9 // 2

print(x, y)
```

```python
x = 12 % 4
y = 9 % 2

print(x, y)
```
