# Słowniki

```python
dct = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4
}

print(dct)
```

```python
dct = dict([('a', 1), ('b', 2), ('c', 3), ('d', 4)])
print(dct)
```

```python
dct = dict.fromkeys(['a', 'b', 'c', 'd', 'e'])
print(dct)
```

```python
dct = dict.fromkeys(['a', 'b', 'c', 'd', 'e'], 'X')
print(dct)
```

```python
dct = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4
}
print(dct['a'])
```

```python
dct = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4
}
print(dct['nie istnieje'])
```

```python
dct = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4
}
print(dct.get('c'))
print(dct.get('nie istnieje'))
print(dct.get('nie istnieje', -1))
```

```python
dct = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4
}
# Nowy element słownika
dct['e'] = 5

print(dct)
```

```python
dct = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4
}
dct.update({'x': 10, 'y': 11, 'a':99})

print(dct)
```

```python
dct = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4
}

v = dct.setdefault('a', -10)
print(v)
print(dct['a'])
v = dct.setdefault('x', -20)
print(v)
print(dct['x'])
v = dct.setdefault('y')
print(v)
print(dct['y'])
```

```python
dct = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4
}
del dct['a']
print(dct)
```
