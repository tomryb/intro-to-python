# Typy danych

## Liczby całkowite

```python
x = 99  # Dzisiętne
y = 0xfeed # Szesnastkowe
z = 0o76 # Ósemkowe

print(x, y, z)
```

## Liczby zmienno przecinkowe

```python
f = 2.
g = .5
h = 2.125

print(f, g, h)
```
